# Training Top Models

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np

# create a dictionary to store models

models = {
    "SVC" : SVC(kernel = 'linear'),   # it is support vector classifier since data is linear as we have seen after plotting the data #for non linear write non-linear instead of linear
                                      # it draws a hyper plane betweeen two classes which lies in the central plane on the basis of nells points of both the classes
    "RandomForest" : RandomForestClassifier(n_estimators = 100, random_state=42), # it trains internally so many decisions i.e 100 as we have set
     
     "GradientBoosting" : GradientBoostingClassifier(n_estimators = 100, random_state=42),
     "KNeighbors" : KNeighborsClassifier(n_neighbors=5),
      "MultinomialNB" : MultinomialNB()
}

# training all 5 models simultaneously

for model_name, model in models.items():
    # train model
    model.fit(x_train, y_train)

    # test model
    predictions = model.predict(x_test)

    # calculate accuracy
    accuracy = accuracy_score(y_test, predictions) #note all models will give 1005 accuracy as data is preprocessed

    # calculate confusion matrix
    cm = confusion_matrix(y_test, predictions) #in confusion matrix more zeros above diagonal indicates high accuracy whereas nonzero value indicates error 

    print(f"Model: {model_name}, Accuracy: {accuracy}")
    print(f"Model: {model_name}, Confusion Matrix:")
    print(np.array2string(cm, separator=', '))
```
